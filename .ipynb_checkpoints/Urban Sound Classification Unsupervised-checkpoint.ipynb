```python
#zca whitening
def zca_whitening_matrix(X):
    # Covariance matrix [column-wise variables]: Sigma = (X-mu)' * (X-mu) / N
    sigma = np.cov(X, rowvar=True) # [M x M]
    # Singular Value Decomposition. X = U * np.diag(S) * V
    U,S,V = np.linalg.svd(sigma)
        # U: [M x M] eigenvectors of sigma.
        # S: [M x 1] eigenvalues of sigma.
        # V: [M x M] transpose of U
    # Whitening constant: prevents division by zero
    epsilon = 1e-5
    # ZCA Whitening matrix: U * Lambda * U'
    ZCAMatrix = np.dot(U, np.dot(np.diag(1.0/np.sqrt(S + epsilon)), U.T)) # [M x M]
    return ZCAMatrix
```

```python
def get_white_mel_spectogram(file_name):
    raw_sound, sample_rate = librosa.load(file_name) #load file
    X = librosa.feature.melspectrogram(y = raw_sound,sr = sample_rate/2,n_fft = 512,hop_length=512,n_mels=n_mel)
    Y = zca_whitening_matrix(X)
    return Y
```

```python
# now first a function to get white_matrix for feature learning
def whitedata(p_dir,s_dirs,file_ext="*.wav"):
    w_data = np.empty(0)
    for label, sub_dir in enumerate(s_dirs):
        for fn in glob.glob(os.path.join(p_dir, sub_dir, file_ext)):
            try:
                X = get_white_mel_spectogram(fn)
                print("Finished processing file: ", fn)
            except Exception as e:
                print("Error while processing file: ", fn)
                print("Exception :", e)
                continue
            try:
                w_data = np.hstack([w_data,X])
            except Exception as e:
                if w_data.shape == (0,):
                    w_data = X
                else:
                    print(w_data.shape,X.shape)

    return np.array(w_data)
```

Next section is feature learning
Objective is to learn a codebook

```python
codebook = np.random.randn(n_mel,k)
S = np.zeros((k,white_matrix.shape[1]),dtype = np.float64)
```

```python
print(white_matrix.shape,white_matrix.dtype)
print(codebook.shape,codebook.dtype)
print(S.shape,S.dtype)
```

```python
def assignsamples():
    global S
    temp = np.dot(white_matrix.transpose(),codebook)
    temp1 = np.argmax(temp,axis=1)
    tempS = np.zeros(S.shape)
    for i in range(S.shape[1]):
        tempS[temp1[i],i] = temp[i,temp1[i]]
    S = tempS
```

```python
def update_centroids():
    global codebook
    codebook = np.dot(white_matrix,S.transpose())+codebook
```

```python
def normalize_codebook():
    global codebook
    for j in range(k):
        codebook[:,j] = codebook[:,j]/np.linalg.norm(codebook[:,j],2)
```

# now we can start with feature extraction

```python
def encodespec(X,code_book):
    #hope this works
    return np.dot(code_book.transpose(),X)
```

```python
def extract_features_from_file(file_name,code_book):
    X =get_white_mel_spectogram(file_name)
    #dimensions of X ar some n*n n=40
    #dimensions of cookbook are n*k
    Y = encodespec(X,code_book)  #function to encode spectogram with codebook
    #dimensions of Y should be then k*m
    
    #Y dimensions will by k number of clusters
    Y_mean = np.mean(Y, axis=1)
    Y_std = np.std(Y, axis=1)
    Y_min = np.min(Y, axis=1)  # row-wise summaries
    Y_max = np.max(Y, axis=1)
    Y_median = np.median(Y, axis=1)
    Y_variance = np.var(Y, axis=1)
    Y_skeweness = skew(Y, axis=1)
    Y_kurtosis = kurtosis(Y, axis=1)
    # pooling functions used are mean and standard deviation
    
    return Y_mean,Y_std,Y_min, Y_max, Y_median, Y_variance, Y_skeweness, Y_kurtosis
```

```python
def extract_features_from_directories(parent_dir, sub_dirs, code_book,file_ext="*.wav"):
    feature_matrix, labels = np.empty((0, featureVectorLength)), np.empty(0)

    for label, sub_dir in enumerate(sub_dirs):
        for fn in glob.glob(os.path.join(parent_dir, sub_dir, file_ext)):
            try:
                Y_mean,Y_std,Y_min, Y_max, Y_median, Y_variance, Y_skeweness, Y_kurtosis = extract_features_from_file(fn,code_book)
                print("Finished processing file: ", fn)
            except Exception as e:
                print("Error while processing file: ", fn)
                print("Exception :", e)
                continue

            # concatenate extracted features
            new_feature_vector = np.hstack([Y_mean,Y_std,Y_min, Y_max, Y_median, Y_variance, Y_skeweness, Y_kurtosis])

            # add current feature vector as last row in feature matrix
            feature_matrix = np.vstack([feature_matrix, new_feature_vector])

            # extracts label from the file name. Change '\\' to  '/' on Unix systems
            labels = np.append(labels, fn.split('\\')[3].split('-')[1])

    return np.array(feature_matrix), np.array(labels, dtype=np.int)
```

```python
training_features, training_labels = extract_features_from_directories(parent_dir, training_dirs,codebook)
test_features, test_labels = extract_features_from_directories(parent_dir, validation_dirs,codebook)
```

```python
def mean_normalize(featureMatrix):
    mean = np.mean(featureMatrix, axis=0)  # compute mean of each column (feature)
    std = np.std(featureMatrix, axis=0, ddof=1)  # compute sample std of each column (feature)

    featureMatrix -= mean  # subtract each column's mean from every value in the corresponding column
    featureMatrix /= std  # divide values in each column with the corresponding sample std for that column

    return featureMatrix
```

```python
import itertools
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt

def accuracy(y_true, y_pred):
    total = len(y_true)
    score = 0
    for i in range(len(y_true)):
        if(y_true[i]==y_pred[i]):
            score += 1
    return (score/total) 


def plot_confusion_matrix(Y_test, Y_pred,cmap=plt.cm.Blues
                          ,title="Confusion Matix"
                          , normalize = False):
   # """
   # This function prints and plots the confusion matrix.
   # Normalization can be applied by setting `normalize=True`.
   # """
   # if normalize:
   #     cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
   #     print("Normalized confusion matrix")
   # else:
   #     print('Confusion matrix, without normalization')

   # print(cm)

    cnf_matrix = confusion_matrix(y_true = Y_test, y_pred = Y_pred)
    class_names = ["air_conditioner","car_horn","children_playing"," dog_bark","drilling","engine_idling"," gun_shot"," jackhammer","siren","street_music"]
    np.set_printoptions(precision=2)

    # Plot non-normalized confusion matrix
    plt.figure(figsize = (10,10))
    
    plt.imshow(cnf_matrix, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(class_names))
    plt.xticks(tick_marks, class_names, rotation=45)
    plt.yticks(tick_marks, class_names)

    fmt = '.2f' if normalize else 'd'
    thresh = cnf_matrix.max() / 2.
    for i, j in itertools.product(range(cnf_matrix.shape[0]), range(cnf_matrix.shape[1])):
        plt.text(j, i, format(cnf_matrix[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cnf_matrix[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
    plt.show()
```
